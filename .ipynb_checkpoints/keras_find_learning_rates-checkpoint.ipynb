```python
import pandas as pd
import numpy as np

from keras.models import *
from keras_lr_finder import LRFinder

x_train = pd.read_csv('trainx.csv')
y_train = pd.read_csv('trainy.csv')
```

```python
model = load_model('glove_rmsprop_compiled.hdf5')
# learning rate finder
lr_finder = LRFinder(model)
lr_finder.find(x_train, y_train, 0.000001, 1, 128, 3)
lr_finder.plot_loss()
```

```python
lr_finder.plot_loss_change(sma=20,n_skip_beginning=20, n_skip_end=5, y_lim=(-0.1, 0.1))
```

```python
history = model.fit(x_train, y_train,
                    batch_size=128,
                    epochs=10,
                    verbose=1,
                    validation_data=xy_val,
                    callback)
```
